# example of how to load data

```python
# modules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# files
#list_of_files = !ls *.txt
list_of_files = !ls ../../pH7.03/CEST/*.txt
n = len(list_of_files)
```

```python
# empty dataframe
CEST_data = np.zeros((101,n))
df = pd.DataFrame(CEST_data)
df.columns = list_of_files
```

```python
for file in list_of_files:
    Z = pd.read_csv(file, header = None)
    df[file] = Z.values[4:] / Z.values[-1]
df[df.columns[-1]] = pd.read_csv('ppm.txt', header = None).values[4:]
df = df.set_index(df.columns[-1]) 
```

```python
df.to_csv('All_data.csv')
```
